Importing the dependencies



```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
#loading the dataset to a pandas dataframe
credit_card_data=pd.read_csv('/content/fraudTrain.csv')
```

```python
credit_card_data.head()
```

```python
credit_card_data.tail()
```

```python
#dataset information
credit_card_data.info()
```

```python
#checking the number of missing values in each column
credit_card_data.isnull().sum()
```

```python
credit_card_data=credit_card_data.dropna(how='any')
```

```python
#distribution of legit transactions & fraudalent transactions
credit_card_data['is_fraud'].value_counts()
```

```python
.36#seperating the data for analysis
legit=credit_card_data[credit_card_data.is_fraud == 0.0]
fraud=credit_card_data[credit_card_data.is_fraud == 1.0]
```

```python
print(legit.shape)
print(fraud.shape)
```

```python
legit.amt.describe()
```

```python
fraud.amt.describe()
```

```python
credit_card_data['trans_date_trans_time'] = pd.to_datetime(credit_card_data['trans_date_trans_time'])
```

```python
#compare the values for both transactions
credit_card_data['amt'].groupby(credit_card_data['is_fraud']).mean()
```

```python
#UNDER SAMPLING

#NO OF FRAUDUALENT TRANSACTIONS=2145
```

```python
legit_sample=legit.sample(n=7506)
```

```python
new_dataset=pd.concat([legit_sample,fraud],axis=0)
```

```python
new_dataset.head()
```

```python
new_dataset.tail()
```

```python
new_dataset['trans_date_trans_time'] = pd.to_datetime(new_dataset['trans_date_trans_time'])
```

```python
new_dataset['is_fraud'].value_counts()
```

```python
new_dataset['amt'].groupby(new_dataset['is_fraud']).mean()
```

```python
x=new_dataset.drop(columns='is_fraud',axis=1)
```

```python
y=new_dataset['is_fraud']
```

```python
print(x)
```

```python
print(y)
```

Spilliting data into training and testing data

```python
x_train,x_test,y_train,y_test = train_test_split(x, y, test_size=0.2,stratify=y,random_state=2)
```

```python
print(x.shape,x_train.shape,x_test.shape)
```

```python
print(y.shape,y_train.shape,y_test.shape)
```

```python
x_train.head()
```

Model training

Logistic Regression


```python
model=LogisticRegression()
```

```python
x_train = x_train[['Unnamed: 0','cc_num','amt','zip','lat','long','city_pop','unix_time']]
```

```python
#training the logistic Regresion model with training data

model.fit(x_train,y_train)
```

Model evalution

Accuracy Score

```python
#accuracy on training data
x_train_prediction=model.predict(x_train)
training_data_accuracy=accuracy_score(x_train_prediction,y_train)
```

```python
print('Accuracy on training data :',training_data_accuracy)
```

```python
x_test = x_test[['Unnamed: 0','cc_num','amt','zip','lat','long','city_pop','unix_time']]
```

```python
#accuracy on test data
x_test_prediction=model.predict(x_test)
test_data_accuracy=accuracy_score(x_test_prediction,y_test)
```

```python
print('Accuracy on test data :',test_data_accuracy)
```
